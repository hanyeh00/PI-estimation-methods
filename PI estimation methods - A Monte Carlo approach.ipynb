```python
#import libraries
import numpy as np
from matplotlib import pyplot as plt
```

```python
#get 2D random points inside a square of size radius
noPoints=10000000 # no. of samples
radius=1.0
samplesX=radius*np.random.rand(noPoints)
samplesY=radius*np.random.rand(noPoints)
```

```python
#check which points fall into the circle part
insideFlag=np.where((samplesX**2+samplesY**2)\
                    <radius**2,1,0)
```

```python
# display a portion of samples and show the PI estimation
points2show=20000
plt.figure(figsize=(5,5))
plt.scatter(samplesX[:points2show],
            samplesY[:points2show],c=insideFlag[:points2show])
plt.xlabel('samplesX'); plt.ylabel('samplesY')
plt.title('pi is almost: %f'%(4*np.sum(insideFlag)/noPoints))
plt.show()
```
